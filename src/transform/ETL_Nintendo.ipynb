O código **nintendo_modificado.groupby('Game').size()** agrupa os dados pela coluna **'Game' (agrupar por jogo)** e conta o tamanho de cada grupo (quantas linhas por jogo). Isso ajuda na limpeza:** identifica duplicatas (jogos com mais de 1 linha)** ou **faltas (jogos com 0 ou poucas entradas)**, permitindo remover incoerências ou preencher dados ausentes.

````python
nintendo_modificado.groupby('Game').size()
````

O código `nintendo_modificado.loc[nintendo_modificado['Game'].isin(['---','----','-----']),'Game']` faz uma filtragem no DataFrame `nintendo_modificado`:

- Procura linhas onde a coluna 'Game' tem **exatamente** um dos valores da lista: '---', '----' ou '-----' (usando `.isin()` para verificar pertencimento exato).
- Retorna apenas a coluna 'Game' dessas linhas (como uma Series), mostrando os nomes filtrados.

````python
nintendo_modificado.loc[nintendo_modificado['Game'].isin(['---','----','-----']),'Game']
````

````python
nintendo_modificado.loc[~nintendo_modificado['Game'].isin(['---','----','-----']),'Game']
````

Se você quer linhas onde 'Game' **contém** (não exato) algo como '---' (ex: strings parciais), use em vez disso `.str.contains(r'---')` no lugar de `.isin()`, ajustando o regex para mais flexibilidade. Exemplo:

```python
nintendo_modificado.loc[nintendo_modificado['Game'].str.contains(r'---+'),'Game']  # captura '---', '----', etc.
```

````python
nintendo_modificado.loc[~nintendo_modificado['Game'].str.contains(r'---+'), 'Game']
````

````python
nintendo_modificado = nintendo_modificado.loc[~nintendo_modificado['Game'].isin(['---','----','-----']),'Game':'Publisher']
nintendo_modificado
````

````python
nintendo_modificado[['Game', 'Platform']] = nintendo_modificado['Game'].str.rsplit(' - ', n=1, expand=True)
nintendo_modificado
````

````python
nintendo_modificado.columns
````

````python
coluna_plataform = nintendo_modificado['Platform']
nintendo_modificado = nintendo_modificado[[ 'Game', 'Platform','Send Date', 'Region', 'Order Status',
                       'E-commerce','Stock Quantity', 'Preço Unit Price',
                        'Store Manager', 'Batch','Age Rating', 'Publisher']]
nintendo_modificado['Platform'] = coluna_plataform
````

````python
nintendo_modificado.columns
````

````python
nintendo_modificado
````

````python
nintendo_modificado.groupby('Region').size()
````

````python
nintendo_modificado['Region'] = nintendo_modificado.loc[:,'Region'].replace({'EN': 'English', 'JP': 'Japanese', 'PT': 'Portuguese'})
````

````python
nintendo_modificado
````

````python
nintendo_modificado['Preço Unit Price'] = nintendo_modificado.loc[:,'Preço Unit Price'].str.replace('$','').astype('float')
````

````python
nintendo_modificado['Stock Quantity'] = nintendo_modificado['Stock Quantity'].astype('int')
````

````python
nintendo_modificado['Stock Quantity']
````

````python
nintendo_modificado['Total Price'] = range(109)
nintendo_modificado = nintendo_modificado[[ 'Game', 'Platform','Send Date', 'Region', 'Order Status',
                       'E-commerce','Stock Quantity', 'Preço Unit Price', 'Total Price',
                        'Store Manager', 'Batch','Age Rating', 'Publisher']]
````

````python
nintendo_modificado['Total Price'] = nintendo_modificado['Stock Quantity'] * nintendo_modificado['Preço Unit Price']
nintendo_modificado['Total Price'] = round(nintendo_modificado['Total Price'],2)
nintendo_modificado
````

````python
nintendo_modificado['Batch'] = nintendo_modificado.loc[:, 'Batch'].str.replace('Batch-','')
nintendo_modificado['Batch']
````

````python
nintendo_modificado['Send Date']
````

````python
nintendo_modificado['Send Date'] = pd.to_datetime(nintendo_modificado['Send Date'], format='%d/%m/%Y', dayfirst=True)
nintendo_modificado['Send Date']
````
