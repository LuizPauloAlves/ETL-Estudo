Pandas Read_html funciona quando tenta extratir informações de tags como 'table', 'tr' e 'td'.
Assim a alternativa é abrir usando BeautifulSoup

```python
from bs4 import BeautifulSoup

image=[]

with open("/content/BS_Pokemon.html") as f:
  soup = BeautifulSoup(f, "html.parser")
for link in soup.find_all("img"):
  data_src = link.get("data-src")
  if data_src and "https://den-cards.pokellector.com/119/" in data_src: #colocar a "proteção" data_src para quando for none não analisar ja que não é iteravel
    image.append(data_src)
for i in range(0,69):
  src = image[i].replace("thumb.","")
  image[i] = src
image[68]
```

```python
import requests
import os

pasta_destino = '/content/pokemon_BS_image'
os.makedirs(pasta_destino, exist_ok=True)
url = 'https://den-cards.pokellector.com/119/Weedle.BS.69.png'

try:
  nome_arquivo = url.split('/')[-1]
  caminho_completo = os.path.join(pasta_destino, nome_arquivo)
  response = requests.get(url, stream=True, timeout=10)
  response.raise_for_status()
  with open(caminho_completo, 'wb') as f:
    for chunk in response.iter_content():
      f.write(chunk)
  print(f"Baixado: {nome_arquivo}")
except Exception as e:
  print(f"Erro ao baixar {url}: {e}")
```

```python
pasta_destino = '/content/pokemon_BS_image'
os.makedirs(pasta_destino, exist_ok=True)

for url in image[0:68]:
    try:
        nome_arquivo = url.split('/')[-1]
        caminho_completo = os.path.join(pasta_destino, nome_arquivo)
        response = requests.get(url, stream=True, timeout=15)
        response.raise_for_status()

        with open(caminho_completo, 'wb') as f:
            for chunk in response.iter_content():
                f.write(chunk)

        print(f"Baixado: {nome_arquivo}")

    except Exception as e:
        print(f"Erro ao baixar {url}: {e}")
```
